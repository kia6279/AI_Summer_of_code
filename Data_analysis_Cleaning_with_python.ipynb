```python
df = df.dropna(axis='columns',how= 'all')
df.head()
```

```python
df = df.drop(columns=['Start Day','End Day', 'Magnitude', 'DisNo.', 'Historic'])
df.head()
```

```python
df.info()
```

```python
cols_to_drop = [0,1,2,3,5,10,11,14,15,16,17,23,25,26,28,29,30]
df.drop(df.columns[cols_to_drop], axis=1, inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
cols_to_fill = ['Start Month','End Month','Total Deaths','Total Affected']
for column_name in cols_to_fill:
  df.loc[df[column_name].isna(), column_name] = 0
df.head()
```

```python
df.info()
```

```python
columns_to_fill = ['Associated Types']
for column_name in columns_to_fill:
  df.loc[df[column_name].isna(), column_name] = 'Unspicified'
df.head()
```

```python
df.info()
```

```python
type_dict = {
    'Start Year':'category',
    'Start Month':'category',
    'End Year':'category',
    'End Month':'category',
    'Total Deaths':'int64',
    'Total Affected':'int64'
}

for keys, vlaues in type_dict.items():
  df[keys] = df[keys].astype(vlaues)

df.head()
```
